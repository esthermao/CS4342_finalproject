```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import kagglehub

# Set plot style for better aesthetics
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)

# Download latest version
path = kagglehub.dataset_download("karkavelrajaj/amazon-sales-dataset")
print("Path to dataset files:", path)

df = pd.read_csv(f"{path}/amazon.csv")

# Display the first 5 rows to get a quick look at the data
df.head()
```

```python
# Get a summary of the DataFrame
df.info()
```

```python
# Get statistical summary for numerical columns
df.describe()
```

```python
df['actual_price'] = df['actual_price'].replace('[₹,]', '', regex=True).astype(float)
df['actual_price'].head()

priceConversion = 0.012 
df['actual_price'].max()
print(f"Maximum actual price: ₹{df['actual_price'].max()} at index {df['actual_price'].idxmax()}")
print(f"Maximum actual price in USD: ${df['actual_price'].max() * priceConversion} at index {df['actual_price'].idxmax()}")
df['actual_price'].min()
print(f"Minimum actual price: ₹{df['actual_price'].min()} at index {df['actual_price'].idxmin()}")
print(f"Minimum actual price in USD: ${df['actual_price'].min() * priceConversion} at index {df['actual_price'].idxmin()}")
df['actual_price'].mean()
print(f"Mean actual price: ₹{df['actual_price'].mean()}")
print(f"Mean actual price in USD: ${df['actual_price'].mean() * priceConversion}")
```

```python
# Plot a histogram of the 'rating' column

# 1. Convert 'rating' column to a numeric type.
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
# 2. Remove any rows with missing ratings
df = df.dropna(subset=['rating'])  # Drop rows where 'rating' is NaN

# Round ratings to 1 decimal place
df['rating_round'] = df['rating'].round(1)

# Count occurrences of each 0.1 bin
rating_counts = df['rating_round'].value_counts().sort_index()

# Ensure all bins 2.5 → 5.0 exist, even if count=0 #because dataset has mostly only has ratings from 2.5 to 5.0
all_bins = np.round(np.arange(2.5, 5.1, 0.1), 1)
rating_counts = rating_counts.reindex(all_bins, fill_value=0)

# Plot as bar chart
plt.figure(figsize=(14,5))
sns.barplot(x=rating_counts.index, y=rating_counts.values, color="skyblue", width=0.9)

plt.title("Distribution of Product Ratings (0.1 bins)", fontsize=18)
plt.xlabel("Rating", fontsize=18)
plt.ylabel("Number of Products", fontsize=18)
plt.xticks(rotation=90, fontsize=18)
plt.yticks(fontsize=18)

plt.tight_layout()
plt.show()
```

```python

# Get the top 10 categories and their counts
top_10_categories = df['category'].value_counts().nlargest(10).index
counts = df['category'].value_counts().nlargest(10)

print(f"Top 10 categories:", top_10_categories.tolist())
print(f"Counts of top 10 categories:", df['category'].value_counts().nlargest(10).values.tolist())


# Extract only the last part of each category
last_categories = [c.split('|')[-1] for c in top_10_categories]

# Create a mapping from full category → last part
category_map = dict(zip(top_10_categories, last_categories))

# Create a new column for last part of category (only for top 10)
df_top_10 = df[df['category'].isin(top_10_categories)].copy()
df_top_10['last_category'] = df_top_10['category'].map(category_map)

# Plot countplot with the simplified labels
sns.countplot(y='last_category', data=df_top_10, order=last_categories)
plt.title('Top 10 Product Categories', fontsize=18)
plt.xlabel('Number of Products', fontsize=14)
plt.ylabel('Category', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
# Create a scatter plot of rating vs. actual_price
sns.scatterplot(x='actual_price', y='rating', data=df, alpha=0.5)
plt.title('Rating vs. Price', fontsize=16)
plt.xlabel('Actual Price', fontsize=12)
plt.ylabel('Rating', fontsize=12)
plt.xscale('log')
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```
