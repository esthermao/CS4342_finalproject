```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import kagglehub

# Set plot style for better aesthetics
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)

# Download latest version
path = kagglehub.dataset_download("karkavelrajaj/amazon-sales-dataset")
print("Path to dataset files:", path)

df = pd.read_csv(f"{path}/amazon.csv")

# Display the first 5 rows to get a quick look at the data
df.head()
```

```python
# Get a summary of the DataFrame
df.info()
```

```python
# Get statistical summary for numerical columns
df.describe()
```

```python
# Plot a histogram of the 'rating' column

# 1. Convert 'rating' column to a numeric type.
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
# 2. Remove any rows with missing ratings
df = df.dropna(subset=['rating'])  # Drop rows where 'rating' is NaN

# Round ratings to 1 decimal place
df['rating_round'] = df['rating'].round(1)

# Count occurrences of each 0.1 bin
rating_counts = df['rating_round'].value_counts().sort_index()

# Ensure all bins 2.5 → 5.0 exist, even if count=0 #because dataset has mostly only has ratings from 2.5 to 5.0
all_bins = np.round(np.arange(2.5, 5.1, 0.1), 1)
rating_counts = rating_counts.reindex(all_bins, fill_value=0)

# Plot as bar chart
plt.figure(figsize=(14,5))
sns.barplot(x=rating_counts.index, y=rating_counts.values, color="skyblue", width=0.9)

plt.title("Distribution of Product Ratings (0.1 bins)")
plt.xlabel("Rating")
plt.ylabel("Number of Products")
plt.xticks(rotation=90, fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Find the top 10 most frequent categories
top_10_categories = df['category'].value_counts().nlargest(10).index

# Filter the DataFrame to include only the top 10 categories
df_top_10 = df[df['category'].isin(top_10_categories)]

# Plot a count plot for the top 10 categories
sns.countplot(y='category', data=df_top_10, order=top_10_categories)
plt.title('Top 10 Product Categories')
plt.xlabel('Number of Products')
plt.ylabel('Category')
plt.show()
```

```python
# Create a scatter plot of rating vs. actual_price
sns.scatterplot(x='actual_price', y='rating', data=df, alpha=0.5)
plt.title('Rating vs. Price')
plt.xlabel('Actual Price')
plt.ylabel('Rating')
plt.xscale('log') # Use a log scale if price is heavily skewed
plt.show()
```
